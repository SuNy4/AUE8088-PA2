```python
import numpy as np
from sklearn.model_selection import KFold

#Enter dataset directory and train set filename
list_dir = "./datasets/kaist-rgbt/"
list_name = "train-all-04.txt"
```

```python
#Open train set txt file and prepare 5 fold
with open(list_dir+list_name, "r") as file:
    data = file.readlines()
kf = KFold(5, shuffle=True, random_state=22)
```

```python
#Create 5 fold data split txt files
fold = 1
print("Split train and val set for 5 fold cross validation")
for train, val in kf.split(data):
    train_data = [data[i] for i in train]
    val_data = [data[i] for i in val]
    train_file = list_dir + f"train_all_04_{fold}.txt"
    val_file = list_dir + f"val_all_04_{fold}.txt"

    with open(train_file, 'w') as file:
        file.writelines(train_data)
        
    with open(val_file, 'w') as file:
        file.writelines(val_data)
    print(f"Complete splitting for fold{fold}")
    fold += 1
```
